```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (replace 'file_path' with the actual path to your file)
file_path = 'API_SP.POP.TOTL_DS2_en_csv_v2_56.csv'
data = pd.read_csv(file_path, skiprows=4)

# Select the year of interest (e.g., 2023)
year = '2023'
population_2023 = data[['Country Name', year]].dropna()

# Bar chart for population distribution (Top 20 countries/regions)
population_top20 = population_2023.sort_values(by=year, ascending=False).head(20)

plt.figure(figsize=(12, 6))
plt.bar(population_top20['Country Name'], population_top20[year], color='purple')
plt.title('Population Distribution by Country (2023)')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Histogram for population values across all countries
plt.figure(figsize=(10, 6))
plt.hist(population_2023[year], bins=15, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Population Values Across Countries (2023)')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```
